```r
poke_df <- read.csv("C:/Users/kgtrm/Documents/VSC Code/R stuff/Pokemon Project/PokemonDataCleaned.csv")
str(poke_df)
```

```r
library(dplyr)
library(gdata)
library(ggplot2)
```

```r
poke_num_df <- poke_df %>%
    select(X, ï..Pokemon.Id, Pokedex.Number, Pokemon.Height, Pokemon.Weight, Male.Ratio, 
                       Female.Ratio, Base.Happiness, Health.Stat, Attack.Stat, Defense.Stat,
                       Special.Attack.Stat, Special.Defense.Stat, Speed.Stat, Base.Stat.Total,
                       Health.EV, Attack.EV, Defense.EV, Special.Attack.EV, Special.Defense.EV,
                       Speed.EV, EV.Yield.Total, Catch.Rate, Experience.Growth.Total, Experience.Yield,
                       Egg.Cycle.Count, Generation)
str(poke_num_df)
```

```r
cor_matrix <- cor(poke_num_df, poke_num_df, method = "pearson")
for (i in 1:27) {
    for (j in 1:27) {
        if (i > j) {
            cor_matrix[i,j] <- 0
        }
    }
}
cor_matrix
```

```r
# assuming results with |cor| >= 0.6 are significant
cor_matrix[abs(cor_matrix) < 0.6] <- 0

cor_matrix
```

```r
high_correlations <- c()
for (i in rownames(cor_matrix)) {
    for (j in colnames(cor_matrix)) {
        if (i != j) {
            if (cor_matrix[i,j] != 0 & cor_matrix[i,j] != 1) {
                high_correlations <- append(high_correlations, c(i,j))
                }  
            }
    }
}

length(high_correlations) / 2
for (i in seq(1, length(high_correlations), 2)) {
    cat(sprintf("\"%s\" \"%s\"\n", high_correlations[i],high_correlations[i + 1]))
}
```
