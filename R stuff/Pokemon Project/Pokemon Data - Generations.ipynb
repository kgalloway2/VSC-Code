```r
library(ggplot2)
library(tidyr)
library(dplyr)
options(repr.plot.width=15, repr.plot.height=8)

poke_df <- read.csv("C:/Users/kgtrm/Documents/VSC Code/R stuff/Pokemon Project/PokemonDataCleaned.csv")
str(poke_df)
```

```r
generations_df <- poke_df %>%
    group_by(Generation) %>%
    summarize(Mean.Health = mean(Health.Stat), Median.Health = median(Health.Stat), MinHealth = min(Health.Stat), MaxHealth = max(Health.Stat), IQRHealth = IQR(Health.Stat),
             Mean.Attack = mean(Attack.Stat), Median.Attack = median(Attack.Stat), MinAttack = min(Attack.Stat), MaxAttack = max(Attack.Stat), IQRAttack = IQR(Attack.Stat),
             Mean.Defense = mean(Defense.Stat), Median.Defense = median(Defense.Stat), MinDefense = min(Defense.Stat), MaxDefense = max(Defense.Stat), IQRDefense = IQR(Special.Defense.Stat),
             Mean.Special.Attack = mean(Special.Attack.Stat), Median.Special.Attack = median(Special.Attack.Stat), MinSpecial.Attack = min(Special.Attack.Stat), MaxSpecial.Attack = max(Special.Attack.Stat), IQRSpecial.Attack = IQR(Attack.Stat),
             Mean.Special.Defense = mean(Special.Defense.Stat), Median.Special.Defense = median(Special.Defense.Stat), MinSpecial.Defense = min(Special.Defense.Stat), MaxSpecial.Defense = max(Special.Defense.Stat), IQRSpecial.Defense = IQR(Special.Defense.Stat),
             Mean.Speed = mean(Speed.Stat), Median.Speed = median(Speed.Stat), MinSpeed = min(Speed.Stat), MaxSpeed = max(Speed.Stat), IQRSpeed = IQR(Speed.Stat),
             Mean.Base.Stat.Total = mean(Base.Stat.Total), Median.Base.Stat.Total = median(Base.Stat.Total), MinBase.Stat.Total = min(Base.Stat.Total), MaxBase.Stat.Total = max(Base.Stat.Total), IQRBase.Stat.Total = IQR(Base.Stat.Total),
             Mean.Experience.Yield = mean(Experience.Yield), Median.Experience.Yield = median(Experience.Yield), MinExperience.Yield = min(Experience.Yield), MaxExperience.Yield = max(Experience.Yield), IQRExperience.Yield = IQR(Experience.Yield),
             Count = n())

generations_df
```

```r
ggplot(data = generations_df) +
    geom_point(aes(x = Generation, y = Mean.Base.Stat.Total)) +
    geom_smooth(aes(x = Generation, y = Mean.Base.Stat.Total), method = "lm", color = "red") +
    geom_point(aes(x = Generation, y = Median.Base.Stat.Total), shape = 17) +
    geom_smooth(aes(x = Generation, y = Median.Base.Stat.Total), method = "lm", color = "blue") +
    ylab(label = "") +
    scale_x_continuous(breaks = seq(1, 8, 1))

ggsave("GenTrendsBST.png")

mean_correlation <- cor(generations_df$Generation, generations_df$Mean.Base.Stat.Total, method = "pearson")
mean_correlation
median_correlation <- cor(generations_df$Generation, generations_df$Median.Base.Stat.Total, method = "pearson")
median_correlation
```

```r
# returns a vector c(value, Stat, Statistic, generation)
mapping_function <- function(datavalue, colname, generation) {
#     if (colname == "Count") {
#         return(c(datavalue, "Count", NA, generation))
#     }
    if (grepl("Mean", colname)) {
        statistic <- "Mean"
    } else if (grepl("Median", colname)) {
        statistic <- "Median"
    } else if (grepl("Min", colname)) {
        statistic <- "Min"
    } else if (grepl("Max", colname)) {
        statistic <- "Max"
    } else if (grepl("IQR", colname)) {
        statistic <- "IQR"
    }
    if (grepl("Health", colname)) {
        stat <- "Health"
    } else if (grepl("Special.Attack", colname)) {
        stat <- "Special Attack"
    } else if (grepl("Special.Defense", colname)) {
        stat <- "Special Defense"
    } else if (grepl("Attack", colname)) {
        stat <- "Attack"
    } else if (grepl("Defense", colname)) {
        stat <- "Defense"
    } else if (grepl("Speed", colname)) {
        stat <- "Speed"
    } else if (grepl("Base", colname)) {
        stat <- "Base Stat Total"
    } else if (grepl("Experience", colname)) {
        stat <- "Experience Yield"
    }
    return(data.frame(Value = datavalue, Stat = stat, Statistic = statistic, Generation = generation))
}
```

```r
# let's reformat our summary data like we wanted it yesterday
generations_df_2 <- data.frame(Value = double(), Stat = character(), Statistic = character(), Generation = integer()) 
for (gen in 1:8) {
    for (col in 2:41) {
        temp_row <- mapping_function(generations_df[[gen, col]], colnames(generations_df)[col], gen)
        # print(temp_row)
        generations_df_2 <- rbind(generations_df_2, temp_row)   
    }  
}
# generations_df_2
```

```r
ggplot(data = generations_df_2) +
    geom_point(aes(x = Generation, y = Value, color = Stat, shape = Statistic))

# it worked but this is A LOT of data, so let's focus on each stat
```

```r
# only want mean and median to keep graphs clear
ggplot(data = generations_df_2[generations_df_2$Statistic == "Mean" | generations_df_2$Statistic == "Median",]) +
    geom_point(aes(x = Generation, y = Value, shape = Statistic)) +
    facet_wrap(~Stat, scales = "free")

ggsave("MeanMedianGenStats.png")
```
