```r
# this is the original data frame
poke_df <- read.csv("C:/Users/kgtrm/Documents/R projects data/archive pokemon better/Pokemon Database.csv")
head(poke_df)
str(poke_df)
```

```r
library(tidyr)
library(dplyr)
```

```r
# this is the raw file for fully evolved pokemon. I am going to extract a list so I can add another attribute to the above table
evolved_raw <- read.delim("C:/Users/kgtrm/Documents/VSC Code/R stuff/Pokemon evolved raw data.txt")
head(evolved_raw)

evolved_split <- evolved_raw %>%
    separate(X...small.Average..small., into = c("temp", "num", "Name"), sep = "\\|") %>%
    select(Name)
head(evolved_split)
str(evolved_split)
```

```r
# now I need this as a list

evolved_list = c()
for (i in 1:nrow(evolved_split)) {
    evolved_list <- append(evolved_list, evolved_split[i,])
}

evolved_list <- unique(evolved_list)
head(evolved_list)
```

```r
# now let's add a column to denote whether a pokemon is fully evolved or not
# note that the Pokemon.Name contains strings with "" included in the string. so we need to remove those
# let's go ahead and do that for all columns where it is needed
poke_df$Pokemon.Name <- gsub("\"", "", poke_df$Pokemon.Name)
poke_df$Classification <- gsub("\"", "", poke_df$Classification)
poke_df$Alternate.Form.Name <- gsub("\"", "", poke_df$Alternate.Form.Name)
poke_df$Experience.Growth <- gsub("\"", "", poke_df$Experience.Growth)
poke_df$Primary.Egg.Group <- gsub("\"", "", poke_df$Primary.Egg.Group)
poke_df$Secondary.Egg.Group <- gsub("\"", "", poke_df$Secondary.Egg.Group)
poke_df$Evolution.Details <- gsub("\"", "", poke_df$Evolution.Details)
poke_df$Primary.Type <- gsub("\"", "", poke_df$Primary.Type)
poke_df$Secondary.Type <- gsub("\"", "", poke_df$Secondary.Type)
poke_df$Primary.Ability <- gsub("\"", "", poke_df$Primary.Ability)
poke_df$Primary.Ability.Description <- gsub("\"", "", poke_df$Primary.Ability.Description)
poke_df$Secondary.Ability <- gsub("\"", "", poke_df$Secondary.Ability)
poke_df$Secondary.Ability.Description <- gsub("\"", "", poke_df$Secondary.Ability.Description)
poke_df$Hidden.Ability <- gsub("\"", "", poke_df$Hidden.Ability)
poke_df$Hidden.Ability.Description <- gsub("\"", "", poke_df$Hidden.Ability.Description)
poke_df$Special.Event.Ability <- gsub("\"", "", poke_df$Special.Event.Ability)
poke_df$Special.Event.Ability.Description <- gsub("\"", "", poke_df$Special.Event.Ability.Description)
poke_df$Game.s..of.Origin <- gsub("\"", "", poke_df$Game.s..of.Origin)
poke_df$Region.of.Origin <- gsub("\"", "", poke_df$Region.of.Origin)

poke_df$Fully.Evolved <- ifelse(poke_df$Pokemon.Name %in% evolved_list, TRUE, FALSE)
head(poke_df, 10)
```

```r
view_games <- poke_df %>%
    select(Game.s..of.Origin)
head(view_games, 20)
```

```r
# now let's add generation for each pokemon
assign_generation <- function(game) {
    if (game == "Red") {
        return(1)
    } else if (game == "Gold") {
        return(2)
    } else if (game == "Ruby") {
        return(3)
    } else if (game == "Fire Red") {
        return(3)
    } else if (game == "Leaf Green") {
        return(3)
    } else if (game == "Emerald") {
        return(3)
    } else if (game == "Diamond") {
        return(4)
    } else if (game == "Platinum") {
        return(4)
    } else if (game == "Black") {
        return(5)
    } else if (game == "Black 2") {
        return(5)
    } else if (game == "X") {
        return(6)
    } else if (game == "Omega Ruby") {
        return(6)
    } else if (game == "Sun") {
        return(7)
    } else if (game == "Ultra Sun") {
        return(7)
    } else if (game == "Let\'s Go Pikachu") {
        return(7)
    } else if (game == "Sword") {
        return(8)
    } else {
        return(NA)
    }
}
```

```r
# now we add a column based on these
poke_df <- poke_df %>%
    rowwise() %>%
    mutate(Generation = assign_generation(Game.s..of.Origin))

head(poke_df)
```

```r
# seems good, now let's save it so we can work with later
write.csv(poke_df, "C:/Users/kgtrm/Documents/VSC Code/R stuff/PokemonDataCleaned.csv")
```
