```r
library(tidyr)
library(dplyr)
library(ggplot2)
options(repr.plot.width=15, repr.plot.height=8)
```

```r
anime_df <- read.csv("C:/Users/kgtrm/Documents/VSC Code/R stuff/Anime Project/AnimeDataCleaned.csv")
head(anime_df)
str(anime_df)
```

```r
# going to inspect the following
#         month ~ rating, 
#         num_ep ~ rating, 
#         types that have risen/fallen in popularity over the past years
```

```r
ggplot(aes(x = factor(Start_Month, levels = c("Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec", NA)), y = Anime_Rating), data = anime_df) +
    geom_point()
```

```r
unique(anime_df$Anime_Type)
```

```r
months_summary <- anime_df %>%
    group_by(Start_Month, Anime_Type) %>%
    summarize(Mean_Rating = mean(Anime_Rating), 
              Median_Rating = median(Anime_Rating),
              Max_Rating = max(Anime_Rating),
              Min_Rating = min(Anime_Rating),
              Mean_Number_Episodes = mean(Number_Episodes, na.rm = TRUE), 
              Median_Number_Episodes = median(Number_Episodes, na.rm = TRUE),
              Number_TV = sum(Anime_Type == "TV "),
              Number_Movies = sum(Anime_Type == "Movie "),
              Number_OVAs = sum(Anime_Type == "OVA "),
              Number_Specials = sum(Anime_Type == "Special "),
              Number_ONAs = sum(Anime_Type == "ONA "),
              Number_Music = sum(Anime_Type == "Music "),
              count = n())

months_summary
```

```r
ggplot(aes(x = factor(Start_Month, levels = c("Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec", NA)), 
           y = Anime_Rating), data = anime_df) +
    geom_point() +
    facet_wrap(~Anime_Type)
```

```r
ggplot(aes(x = factor(Start_Month, levels = c("Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec", NA)), 
           y = Mean_Rating), data = months_summary) +
    geom_point()+
    xlab(label = "Release Month") +
    facet_wrap(~Anime_Type)
```

```r
# no notable conclusions other than 
# TV anime released in the four main months enjoy higher ratings on average than the other months
# Movies released in the months feb, apr, aug, sep enjoy higher ratings on average than others

# similar conclusions for medians below
```

```r
ggplot(aes(x = factor(Start_Month, levels = c("Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec", NA)), 
           y = Median_Rating), data = months_summary) +
    geom_point()+
    xlab(label = "Release Month")+
    facet_wrap(~Anime_Type)
```

```r
TV_only = anime_df[which(anime_df$Anime_Type == "TV "),]

ggplot(aes(x = Number_Episodes, y = Anime_Rating), data = TV_only) +
    geom_point() +
    geom_smooth()
```

```r
cor(x = TV_only$Number_Episodes, y = TV_only$Anime_Rating, method = "pearson", use = "complete.obs")
```

```r
# insignificant correlation even though the graph appears to show an inverse relationship
```

```r
years_summary <- anime_df %>%
    group_by(Start_Year, Anime_Type) %>%
    summarize(Mean_Rating = mean(Anime_Rating), 
              Median_Rating = median(Anime_Rating),
              Max_Rating = max(Anime_Rating),
              Min_Rating = min(Anime_Rating),
              Mean_Number_Episodes = mean(Number_Episodes, na.rm = TRUE), 
              Median_Number_Episodes = median(Number_Episodes, na.rm = TRUE),
              Number_TV = sum(Anime_Type == "TV "),
              Number_Movies = sum(Anime_Type == "Movie "),
              Number_OVAs = sum(Anime_Type == "OVA "),
              Number_Specials = sum(Anime_Type == "Special "),
              Number_ONAs = sum(Anime_Type == "ONA "),
              Number_Music = sum(Anime_Type == "Music "),
              count = n())

years_summary
```

```r
# trends of mean ratings over years by type
ggplot(aes(x = Start_Year, y = Mean_Rating), data = years_summary)+
    geom_point() +
    facet_wrap(~Anime_Type, scales = "free")
```

```r
# trends but for median ratings
ggplot(aes(x = Start_Year, y = Median_Rating), data = years_summary)+
    geom_point() +
    geom_smooth() +
    facet_wrap(~Anime_Type, scales = "free")
```

```r
# seems like everything has gotten more popular over the years
# perhaps a decline in tv and specials over the past couple years
```

```r
# what if we only look at things with ratings above 8.5
above_9s <- anime_df[which(anime_df$Anime_Rating >= 8.5),]
```

```r
above_9s_months_summary <- above_9s %>%
    group_by(Start_Month, Anime_Type) %>%
    summarize(Mean_Rating = mean(Anime_Rating), 
              Median_Rating = median(Anime_Rating),
              Max_Rating = max(Anime_Rating),
              Min_Rating = min(Anime_Rating),
              Mean_Number_Episodes = mean(Number_Episodes, na.rm = TRUE), 
              Median_Number_Episodes = median(Number_Episodes, na.rm = TRUE),
              Number_TV = sum(Anime_Type == "TV "),
              Number_Movies = sum(Anime_Type == "Movie "),
              Number_OVAs = sum(Anime_Type == "OVA "),
              Number_Specials = sum(Anime_Type == "Special "),
              Number_ONAs = sum(Anime_Type == "ONA "),
              Number_Music = sum(Anime_Type == "Music "),
              count = n())

# above_9s_months_summary


ggplot(aes(x = factor(Start_Month, levels = c("Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec", NA)), 
           y = Mean_Rating), data = above_9s_months_summary) +
    geom_point()+
    xlab(label = "Release Month")+
    facet_wrap(~Anime_Type)
```

```r
above_9s_TV_only = above_9s[which(above_9s$Anime_Type == "TV "),]

ggplot(aes(x = Number_Episodes, y = Anime_Rating), data = above_9s_TV_only) +
    geom_point() +
    geom_smooth()
```

```r
above_9s_years_summary <- above_9s %>%
    group_by(Start_Year, Anime_Type) %>%
    summarize(Mean_Rating = mean(Anime_Rating), 
              Median_Rating = median(Anime_Rating),
              Max_Rating = max(Anime_Rating),
              Min_Rating = min(Anime_Rating),
              Mean_Number_Episodes = mean(Number_Episodes, na.rm = TRUE), 
              Median_Number_Episodes = median(Number_Episodes, na.rm = TRUE),
              Number_TV = sum(Anime_Type == "TV "),
              Number_Movies = sum(Anime_Type == "Movie "),
              Number_OVAs = sum(Anime_Type == "OVA "),
              Number_Specials = sum(Anime_Type == "Special "),
              Number_ONAs = sum(Anime_Type == "ONA "),
              Number_Music = sum(Anime_Type == "Music "),
              count = n())

# above_9s_years_summary

ggplot(aes(x = Start_Year, y = Mean_Rating), data = above_9s_years_summary)+
    geom_point() +
    facet_wrap(~Anime_Type, scales = "free")
```
