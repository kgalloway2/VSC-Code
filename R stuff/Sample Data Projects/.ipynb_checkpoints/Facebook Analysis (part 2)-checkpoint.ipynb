```r
# aes wrapper
# histogram of age vs friend count
ggplot(aes(x = age, y = friend_count), data = df_fb) +
    geom_point(alpha = 0.05) +
    xlim(13, 90)
```

```r
# we can use jitter to reduce overplotting
ggplot(aes(x = age, y = friend_count), data = df_fb) +
    geom_jitter(alpha = 0.05) + 
    xlim(13, 90)
```

```r
# scatterplot of age vs friend count
ggplot(aes(x = age, y = friend_count), data = df_fb) +
    geom_point(alpha = 0.05) +
    xlim(13, 90) +
    coord_trans(y = "sqrt") # transforms the y-coordinates to sqrts of themselves
```

```r
ggplot(aes(x = age, y = friend_count), data = df_fb) +
    geom_point(alpha = 0.05,
              position = position_jitter(h = 0),
              color = "red") +
    xlim(13, 90) +
    coord_trans(y = "sqrt")
```

```r
# adding dplyr library
library(dplyr)
```

```r
# bin ages, create new dataframe with mean and median friend coutns for each age group
# copy of df_fb that is grouped by age
age_groups <- group_by(df_fb, age)
# make summary frame of that
df_fb_by_age <- summarize(age_groups,
                         mean_friend_count = mean(friend_count),
                         median_friend_count = median(friend_count),
                         n = n())
# arrange it
df_fb_by_age <- arrange(df_fb_by_age, age)
head(df_fb_by_age, 15)
```

```r
# scatterplot of mean friend counts for each age group
ggplot(aes(x = age, y = mean_friend_count), data = df_fb_by_age) +
    geom_line() +
    xlim(13, 90)
```

```r
# scatterplot of age vs friend count, using a jitter to reduce overplotting
ggplot(aes(x = age, y = friend_count), data = df_fb) +
    geom_point(alpha = 0.05,
              position = position_jitter(h = 0),
              color = "red")+
    xlim(13, 90)+
    coord_trans(y = "sqrt") +
# so far, all stuff we have done before
# now let's add some statistics
    geom_line(stat = "summary", fun = mean, color = "yellow") +
    geom_line(stat = "summary", fun = quantile, fun.args = list(probs = 0.2), color = "blue") +
    geom_line(stat = "summary", fun = quantile, fun.args = list(probs = 0.8), color = "green") +
    geom_line(stat = "summary", fun = median, color = "black")
```

```r
# correlation coefficient between age and friend count using
# pearson correlation coefficient
cor.test(df_fb$age, df_fb$friend_count, method = "pearson")
# OR
with(df_fb, cor.test(age, friend_count, metho = "pearson"))
```

```r
#subset the data frame for age < 70 and look at cor coeff again
with(subset(df_fb, age < 70), cor.test(age, friend_count, method = "pearson"))
```
